```python
# teamgapê³¼ winteamì˜ ì¡°í•©ë³„ ë¹ˆë„ìˆ˜ ê³„ì‚°
value_counts = df3.groupby(['teamgap', 'winteam']).size().unstack(fill_value=0)

# Seabornì„ ì‚¬ìš©í•˜ì—¬ íŒŒì´ ì°¨íŠ¸ ê·¸ë¦¬ê¸°
sns.set(style="whitegrid")
fig, axes = plt.subplots(1, len(value_counts), figsize=(15, 5))

for idx, (gap, row) in enumerate(value_counts.iterrows()):
    sns.barplot(x=row.index, y=row.values, ax=axes[idx])
    axes[idx].set_title(f'Teamgap: {gap}')

plt.tight_layout()
plt.show()
```
